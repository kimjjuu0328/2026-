```python
import numpy as np

arr =  np.array([1,2,3]) #파이썬 리스트 타입을 Numpy의 ndarray형식으로 생성
print(type(arr))
print(arr)
```

```python
import numpy as np

arr =  np.array([[1,2,3],[4,5,6]]) #파이썬 리스트 타입을 Numpy의 ndarray형식으로 생성
print(arr.ndim)
print(arr.shape)
```

```python
import numpy as np

arr =  np.arange(5) #Numpy의 ndarray를 0부터 4까지 1씩 증가시켜 생성
print(arr)
arr =  np.arange(5,15,2) #Numpy의 ndarray를 5부터 14까지 2씩 증가시켜 생성
print(arr)
```
