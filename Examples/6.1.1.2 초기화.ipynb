```python
import numpy as np

arr =  np.zeros(5) #Numpyмқҳ ndarrayмқ„ 0мңјлЎң мұ„мӣҢ мғқм„ұ
print(arr)
```

```python
import numpy as np

arr =  np.zeros((2,5)) #Numpyмқҳ ndarrayмқ„ 0мңјлЎң мұ„мӣҢ 2,5 нҳ•нғңлЎң мғқм„ұ
print(arr)
arr =  np.zeros((2,2,2)) #Numpyмқҳ ndarrayмқ„ 0мңјлЎң мұ„мӣҢ 2,2,2 нҳ•нғңлЎң мғқм„ұ
print(arr)
```

```python
import numpy as np

arr =  np.ones(5) #Numpyмқҳ ndarrayмқ„ 1лЎң мұ„мӣҢ мғқм„ұ
print(arr)
arr =  np.ones((2,2)) #Numpyмқҳ ndarrayмқ„ 1лЎң мұ„мӣҢ  2,2нҳ•нғңлЎң мғқм„ұ
print(arr)
```

```python
import numpy as np

arr =  np.full(5,-1) #Numpyмқҳ ndarrayмқ„ -1лЎң мұ„мӣҢ мғқм„ұ
print(arr)
arr =  np.full((2,2),-1) #Numpyмқҳ ndarrayмқ„ -1лЎң мұ„мӣҢ 2,2нҳ•нғңлЎң мғқм„ұ
print(arr)
```

```python
import numpy as np

arr =  np.eye(5) #Numpyмқҳ ndarrayмқ„ лҢҖк°Ғм„ мңјлЎң 1лЎң мұ„мҡ°кі  лӮҳлЁём§ҖлҠ” 0мңјлЎң мұ„мӣҢ 5,5 нҳ•нғңлЎң мғқм„ұ
print(arr)
```

```python
import numpy as np

arr =  np.random.random(5) #Numpyмқҳ ndarrayмқ„ лһңлҚӨ к°’мңјлЎң мұ„мӣҢ мғқм„ұ
print(arr)
arr =  np.random.random((2,2)) #Numpyмқҳ ndarrayмқ„ лһңлҚӨ к°’мңјлЎң мұ„мӣҢ 2,2 нҳ•нғңлЎң мғқм„ұ
print(arr)
```
