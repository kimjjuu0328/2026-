```python
from popAssist import *
import time

stream = create_conversation_stream()
ga = GAssistant(stream)

try:
    def onStart():
        print(">>> start recording....")

    def onStop():
        ga.assist(onStart, onStop)  #응답의 출력이 완료되면 다시 assist() 호출

    ga.assist(onStart, onStop)
    while True: 
        print("main work...")
        time.sleep(2)
except KeyboardInterrupt:
    stream.close()
```
