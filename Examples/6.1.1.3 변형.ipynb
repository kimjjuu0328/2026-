```python
import numpy as np

arr =  np.arange(10) #Numpy의 ndarray을 0 부터 10까지로 채워 생성
print(arr)
reshp =  arr.reshape(2,5) #arr을 2,5 형태로 재구성
print(reshp)
```

```python
import numpy as np

arr =  np.arange(10) #Numpy의 ndarray을 0 부터 10까지로 채워 생성
print(arr)
s1 =  arr[5] #arr에서 5번째 값 추출
print(s1)
s2 = arr[:5] #arr에서 1 ~ 5번째 값 추출
print(s2)
s3 = arr[2:5] #arr에서 3 ~ 5번째 값 추출
print(s3)
```

```python
import numpy as np

arr =  np.arange(10).reshape(2,5) #Numpy의 ndarray을 0 부터 10까지로 채워 생성하고 2,5 형태로 변형
print(arr)
s1 =  arr[1,4] #arr에서 2행 5열 값 추출
print(s1)
s2 = arr[1:] #arr에서 2행 부터 끝행까지 추출
print(s2)
s3 = arr[0, :] #arr에서 1행의 모든 열 값 추출
print(s3)
s4 = arr[:, 0] #arr에서 모든 행의 1 열 값 추출
print(s4)
```
