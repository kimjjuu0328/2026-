```python
import pyaudio
import wave
import time

CHUNK = 1024
RATE = 48000
isStop = False
```

```python
p = pyaudio.PyAudio()

w = wave.open("out.wav", "wb")
w.setsampwidth(p.get_sample_size(pyaudio.paInt16))
w.setnchannels(1)
w.setframerate(RATE)
```

```python
def callback(in_data, frame_count, time_info, status):
    w.writeframes(in_data)
    data = chr(0) * len(in_data)
    return (data, pyaudio.paContinue if not isStop else pyaudio. paComplete)
```

```python
stream = p.open(format=pyaudio.paInt16, 
                channels=1, rate=RATE, input=True, 
                frames_per_buffer=CHUNK, 
                stream_callback=callback)
```

```python
stream.start_stream()
try:
    print("Recording..", end="")
    while True:
        print(".", end="")
        time.sleep(0.5)
except KeyboardInterrupt:
    isStop = True
    time.sleep(1)   # 녹음을 완료할 때까지 대기
```

```python
w.close()
stream.stop_stream()
stream.close()
p.terminate()
```
