```python
import matplotlib.pyplot as plt
import numpy as np

data=np.random.random(10) #10개의 랜덤 값을 가진 1차원 ndarray 생성
plt.bar(np.arange(10), data) #막대 그래프로 출력
plt.show()
```

```python
plt.bar(np.arange(10), data, width=0.1) #얇은 막대 그래프로 출력
plt.show()
plt.bar(np.arange(10), data, align="edge") #모서리리 정렬 막대 그래프로 출력
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data=np.random.random(30).reshape(3,10) #30개의 랜덤 값을 가진 1차원 ndarray 생성하고 3,10 형태로 변환
plt.bar(np.arange(10), data[0], color="r") #빨간색 막대 그래프로 출력
plt.bar(np.arange(10), data[1], color="g") #초록색 막대 그래프로 출력
plt.bar(np.arange(10), data[2], color="b") #파란색 막대 그래프로 출력
plt.show()
```

```python
data=np.random.random(15).reshape(3,5)
plt.bar(np.arange(0,25,5), data[0], color="r") #빨간색 막대 그래프로 출력
plt.bar(np.arange(0,25,5)+1, data[1], color="g") #초록색 막대 그래프를 1칸 띄워 출력
plt.bar(np.arange(0,25,5)+2, data[2], color="b") #파란색 막대 그래프를 2칸 띄워 출력
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data=np.random.random(20).reshape(2,10) #20개의 랜덤 값을 가진 1차원 ndarray 생성하고 2,10 형태로 변환
plt.bar(np.arange(10), data[0], color="r") #빨간색 막대 그래프로 출력
plt.bar(np.arange(10), -data[1], color="b") #파란색 막대 그래프로 출력
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data=np.random.random(20).reshape(2,10) #20개의 랜덤 값을 가진 1차원 ndarray 생성하고 2,10 형태로 변환
plt.bar(np.arange(10), data[0], color="r") #빨간색 막대 그래프로 출력
plt.bar(np.arange(10), data[1], color="b", bottom=data[0]) #파란색 막대 그래프로 출력
plt.show()
```
