```python
import pandas as pd

value = [[32, 68, 220, 72],
	[28, 30, 0, 12],
	[38, 81, 0, 91]]
columns = ["온도", "습도", "강수량", "불쾌지수"]
index = ["초여름", "늦봄", "한여름"]
df = pd.DataFrame(value, index=index, columns=columns)

print(df.drop("불쾌지수", 1)) #데이터 프레임에서 "불쾌지수" 열을 삭제하여 반환
```

```python
print(df.drop("초여름", 0)) #데이터 프레임에서 "초여름" 행을 삭제하여 반환
```
