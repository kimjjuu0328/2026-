```python
import pandas as pd

value = [[32, 68, 220, 72],
	[28, 30, 0, 12],
	[38, 81, 0, 91]]
df = pd.DataFrame(value)

print(df) #데이터 프레임을 반환
print(df[0][1]) #데이터 프레임의 0번째 행 중 1번째 열을 반환
```

```python
import pandas as pd

value = [[32, 68, 220, 72],
	[28, 30, 0, 12],
	[38, 81, 0, 91]]
columns = ["온도", "습도", "강수량", "불쾌지수"]
index = ["초여름", "늦봄", "한여름"]
df = pd.DataFrame(value, index=index, columns=columns)

print(df) #데이터 프레임을 반환
print(df["온도"]["늦봄"]) #데이터 프레임에서 "온도" 행의 "늦봄" 열 반환
```

```python
import pandas as pd

value = [[32, 68, 220, 72],
	[28, 30, 0, 12],
	[38, 81, 0, 91]]
columns = ["온도", "습도", "강수량", "불쾌지수"]
index = ["초여름", "늦봄", "한여름"]
df = pd.DataFrame(value, index=index, columns=columns)

print(df.index) #데이터 프레임의 인덱스를 반환
print(df.columns) #데이터 프레임의 열을 반환
print(df.values) #데이터 프레임의 값을 반환
```
