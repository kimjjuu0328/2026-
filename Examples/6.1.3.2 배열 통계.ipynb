```python
import numpy as np

x =  np.arange(4).reshape(2,2) #Numpy의 ndarray을 0 부터 3까지로 채워 생성하고 2,2 형태로 변형
print(x)

s1 =  np.sum(x) #ndarray x 원소들의 합을 반환
print(s1)
s2 =  np.mean(x) #ndarray x 원소들의 평균을 반환
print(s2)
s3 =  np.max(x) #ndarray x 원소들 중 최댓값을 반환
print(s3)
s4 =  np.min(x) #ndarray x 원소들 중 최솟값을 반환
print(s4)
```

```python
import numpy as np

x =  np.arange(4).reshape(2,2) #Numpy의 ndarray을 0 부터 3까지로 채워 생성하고 2,2 형태로 변형
print(x)

s1 =  np.std(x) #ndarray x 원소들의 표준편차를 반환
print(s1)
s2 =  np.var(x) #ndarray x 원소들의 분산을 반환
print(s2)
s3 =  np.cumsum(x) #ndarray x 원소들의 누적 합을 반환
print(s3)
s4 =  np.cumprod(x) #ndarray x 원소들의 누적 곱을 반환
print(s4)
```
