```python
import matplotlib.pyplot as plt
import numpy as np

data=np.random.random(30).reshape(10,3) #30개 랜덤 값을 가진 배열 생성 후 10,3 형태의 2차원 배열로 변환
plt.title("stock prices") #그래프에 대한 제목 지정
plt.xlabel("time") #그래프의 X축에 대한 라벨 지정
plt.ylabel("dollar") #그래프의 Y축에 대한 라벨 지정
plt.plot(data) #선형 그래프로 출력
plt.legend(["None-pi","Panthers","Macplotlib"]) #각 그래프에 대한 범례 지정
plt.show()
```
