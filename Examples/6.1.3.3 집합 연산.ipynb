```python
import numpy as np

x =  np.ones((2,2)) #Numpy의 ndarray을 1로 채워 2,2형태로 생성
print(x)
y =  np.arange(3,-1,-1).reshape(2,2) #Numpy의 ndarray을 3 부터 0까지로 채워 생성하고 2,2 형태로 변형
print(y)

s1 =  np.sort(y) #ndarray y의 각 행을 오름차순으로 정렬하여 반환
print(s1)
s2 =  np.unique(x) #ndarray x 원소들의 중복을 제거하여 반환
print(s2)
s3 =  np.union1d(x,y) #ndarray x 와 y의 합집합을 반환
print(s3)
s4 =  np.setdiff1d(y,x) #ndarray y 에 대한 x의 차집합을 반환
print(s4)
```
