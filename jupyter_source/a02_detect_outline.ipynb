- autocar 를 불러와서 카메라를 킨다

```python
import cv2   # 4.3 버전
from pop import Util

Util.enable_imshow()
```

```python
cam = Util.gstrmer()
```

```python
cap = cv2.VideoCapture(cam, cv2.CAP_GSTREAMER)
if not cap.isOpened():
    print("Not found camera")
try:
    for _ in range(12000):
        ret, frame = cap.read()
        if not ret:
            break
        img = cv2.Canny(frame, 100, 200)
        cv2.imshow("soda", img)
except KeyboardInterrupt:
    pass
finally:
    cap.release()
# cv2.destroyAllWindows()
```
