```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(np.array(LR.X_data), np.array(LR.Y_data))
print(model.coef_)
print(model.intercept_)
```

```python
import tensorflow as tf
model= tf.keras.Sequential([tf.keras.layers.Dense(1, input_shape=[1])])
```

```python
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.00001), loss="mse")
```

```python
weight, bias = model.layers[0].get_weights()
print(weight, bias)
model.fit(np.array(LR.X_data), np.array(LR.Y_data), epochs=100, verbose=0)
weight, bias = model.layers[0].get_weights()
print(weight, bias)
```

```python
print("x=6 예측:", model.predict([[173]]))
```
