```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

W1 = tf.Variable(tf.random_uniform([1,10], -1., 1.))
W2 = tf.Variable(tf.random_uniform([10,10], -1., 1.))
W3 = tf.Variable(tf.random_uniform([10,1], -1., 1.))
```

```python
L1 = tf.matmul(X, W1)
L1 = tf.nn.relu(L1)
L2 = tf.matmul(L1, W2)
L2 = tf.nn.relu(L2)
model = tf.matmul(L2, W3)
Loss = tf.reduce_mean(tf.reduce_mean(tf.square(Y - model)))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.001)
train_op = optimizer.minimize(Loss)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
import numpy as np
X_data = np.array([-4,-3, -2, -1, 0, 1, 2, 3, 4]).reshape(-1, 1)
Y_data = np.array([-8, -6, -4, -2, 0, 2, 4, 6, 8]).reshape(-1, 1)

for i in range(1000):
    sess.run(train_op, feed_dict={X: X_data, Y: Y_data})
    loss = sess.run(Loss, feed_dict={X: X_data, Y: Y_data})
    print(loss)
```
