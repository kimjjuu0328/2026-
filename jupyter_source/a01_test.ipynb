```python
import cv2

image = cv2.imread("img.jpg")
height, width, channel = image.shape

print(f"width {width}, height {height}, channel {channel}")
image2 = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
canny = cv2.Canny(image2, 100, 200)
```

```python
cv2.imshow("image", image)
cv2.imshow("image2", image2)
cv2.imshow("canny", canny)
```

```python
image = cv2.imread("img.jpg", cv2.IMREAD_GRAYSCALE)
height, width = image.shape

print(f"width {width}, height {height}")
cv2.imshow("image", image)
```

```python
b, g, r = cv2.split(image)   # opencv 에서는 생삭 순서를 bgr로 한다
imgNew = cv2.merge((r,g,b))  # 일상생활에서 쓰이는 순서
cv2.imshow("imagergb", imgNew)
cv2.imwrite("img_rgb.bmp", imgNew)
```
