```python
import wave

import pyaudio

p = pyaudio.PyAudio()
with wave.open("/usr/share/sounds/alsa/Side_Left.wav", "rb") as w:
    data = w.readframes(w.getnframes())
    stream = p.open(format=p.get_format_from_width(2), channels=1, rate=48000, output=True)
    print("재생 시작")
    stream.write(data)     # 블럭킹
    stream.stop_stream()
    print("재생 끝")
    stream.close()
```
