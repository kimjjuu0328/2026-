```python
dnn.train()
```

```python
import numpy as np
result = dnn.run()
pred_index = np.argmax(result, axis=1)
true_index = np.argmax(dnn.Y_data, axis=1)
classes = ["여성", "남성"]
accuracy = np.mean(pred_index == true_index)
print(f"정확도는 : {accuracy}")
```
