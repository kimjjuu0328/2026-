```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

C = tf.constant(10)
V = tf.Variable(5)

print(C)
print(V)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())

print(sess.run(C))
print(sess.run(V))
```

```python
C = tf.constant(10)
V = tf.Variable(5)
X = tf.placeholder(tf.int32)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

F1 = C * X
F2 = V * X

R1 = sess.run(F1, feed_dict = {X : 2})
R2 = sess.run(F2, feed_dict = {X : 5})

print(R1)
print(R2)
```
