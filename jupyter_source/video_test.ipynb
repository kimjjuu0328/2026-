```python
from pop import Util
import cv2

cam = Util.gstrmer(width=640, height=480, fps=30, flip=0)
cap = cv2.VideoCapture(cam, cv2.CAP_GSTREAMER)
Util.enable_imshow()

for _ in range(120):
    ret, frame = cap.read()
    if not ret:
        print(ret)
        continue
    cv2.imshow("frame", frame)
cap.release()
```
