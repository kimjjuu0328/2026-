```python
import gym
from pop import Util

env = gym.make("CartPole-v1")
obs = env.reset()

for _ in range(1000):
    img = env.render(mode="rgb_array")
    Util.imshow("CartPole", img, mode="RGB")

    action = env.action_space.sample()
    obs, reward, done, info = env.step(action)

    if done:
        obs = env.reset()

env.close()
```
