```python
from pop import AI
perc = AI.Perceptron(input_size=3)
```

```python
import numpy as np

a1 = [173, 171, 162, 187, 157, 169, 177, 159, 182]
a2 = [270, 275, 245, 280, 230, 265, 270, 250, 275]
a3 = [17, 51, 62, 12, 47, 30, 5, 32, 0]
a4 = [0, 1, 1, 0, 1, 0, 0, 1, 0]

perc.X_data = np.array([a1,a2,a3], dtype=np.float32).T
perc.Y_data = np.array(a4, dtype=np.float32).reshape(-1,1)
```

```python
perc.train()
```

```python
perc.run()
```

```python
perc.train(times=1000, print_every=100)
```

```python
pred = perc.run()
print(pred)
pred_label = (pred >= 0.5).astype(np.float32)
accuracy = np.mean(pred_label.reshape(-1,1) == perc.Y_data)
print(accuracy)
```
