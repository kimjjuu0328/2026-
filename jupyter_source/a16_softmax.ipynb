```python
import numpy as np

a1 = [73, 62, 83, 110, 139, 123, 177, 159, 182]
a2 = [1,1,1,0,0,0,0,0,0]
a3 = [0,0,0,1,1,1,0,0,0]
a4 = [0,0,0,0,0,0,1,1,1]
```

```python
import tensorflow as tf

X = np.array(a1, dtype=np.float32).reshape(-1,1)
Y = np.array([a2, a3, a4], dtype=np.float32).T

classes = ["유아", "어린이", "성인"]

X_mean = X.mean()  # normalization 
X_std = X.std()

X_scaled = (X - X_mean) / X_std

model = tf.keras.Sequential(tf.keras.layers.Dense(3, input_shape=[1], activation="softmax"))

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.05),
             loss="categorical_crossentropy",
             metrics=["accuracy"])
```

```python
print(tf.__version__)
```

```python
model.fit(X_scaled, Y, epochs=10, verbose=0)
```

```python
test = [70, 120, 170]
Test = np.array(test, dtype=np.float32).reshape(-1,1)
Test_scaled = (Test - X_mean) / X_std
pred = model.predict(Test_scaled)

for height, prob in zip(Test.flatten(), pred):
    print(f"\n키: {height}cm")
    
    for class_name, p in zip(classes, prob):
        print(f"{class_name}: {p:.4f}")
    result = classes[np.argmax(prob)]
    print("최종 예측: ", result)
```
