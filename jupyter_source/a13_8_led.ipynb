```python
from pop import PwmController
import time

pwm = PwmController(0x57)
pwm.init()
pwm.setChannel(-1)
pwm.setFreq(50)
pwm.setDuty(0)

for i in range(30):
    if i % 2:
        pwm.setDuty(0)
    else:
        pwm.setDuty(100)
    time.sleep(0.5)
```

```python
import time
from pop import Input
from IPython.display import display
from ipywidgets import widgets
import Jetson.GPIO as GPIO

SW1_PIN = 16
SW2_PIN = 13

count = 0

def button_cb(name):
    global count
    count += 1
    print(name, "버튼이 눌렸어요", count)

try:
    GPIO.remove_event_detect(SW2_PIN)
except:
    pass

sw1 = Input(SW1_PIN)
sw2 = Input(SW2_PIN)

sw2.setCallback(lambda channel: button_cb("SW2"), type=GPIO.BOTH)

SW1 = widgets.Label(value="SW1 : 0")
SW2 = widgets.Label(value="SW2 : 0")

display(SW1)
display(SW2)

start_time = time.time()

while time.time() - start_time < 30:
    SW1.value = "SW1 : " + str(sw1.read())
    SW2.value = "SW2 : " + str(sw2.read()) + " / count : " + str(count)
    time.sleep(0.05)
```

```python
del sw1, sw2
```
